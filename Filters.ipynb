```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread('image.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

# Blue

```python
def increase_blue(image,increaseing_value):
    r,g,b = cv2.split(image)
    limit = 255 - increaseing_value
    b[b>limit] = 255
    b[b<limit] += increaseing_value
    image = cv2.merge((r,g,b))
    return image

def decrease_blue(image,decreasing_value):
    r,g,b = cv2.split(image)
    limit = decreasing_value
    b[b<limit] = 0
    b[b>limit] -= decreasing_value
    image= cv2.merge((r,g,b))
    return image
```

```python
def increase_red(image,increaseing_value):
    r,g,b = cv2.split(image)
    limit = 255 - increaseing_value
    r[r>limit] = 255
    r[r<limit] += increaseing_value
    image= cv2.merge((r,g,b))
    return image 

def decrease_red(image,decreasing_value):
    r,g,b = cv2.split(image)
    limit = decreasing_value
    r[r<limit] = 0
    r[r>limit] -= decreasing_value
    image= cv2.merge((r,g,b))
    return image
```

```python
def increase_green(image,increaseing_value):
    r,g,b = cv2.split(image)
    limit = 255 - increaseing_value
    g[g>limit] = 255
    g[g<limit] += increaseing_value
    
    image= cv2.merge((r,g,b))
    return image

def decrease_green(image,decreasing_value):
    r,g,b = cv2.split(image)
    limit = decreasing_value
    g[g<limit] = 0
    g[g>limit] -= decreasing_value
    image= cv2.merge((r,g,b))
    return image
```

```python
def Lord_Kelvin(image):
    image=increase_red(image,110)
    image=increase_green(image,40)
    plt.imshow(image)
```

```python
def Charming(image):
    image=increase_red(image,80)
    image=increase_blue(image,60)
    plt.imshow(image)
```

```python
def Forest(image):
    image=decrease_blue(image,5)
    image=increase_green(image,25)
    image=decrease_red(image,20)
    plt.imshow(image)
```

```python
def Hot_Desert(image):
    image=increase_red(image,40)
    image=decrease_blue(image,100)
    image=decrease_green(image,30)
    plt.imshow(image)
```

```python
def Gray(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
```
